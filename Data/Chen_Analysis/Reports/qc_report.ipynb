```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../CheckM2/quality_report.tsv", sep="\t")
df.head()
```

```python
h_q = df[(df["Completeness"] >= 90) & (df["Contamination"] < 5)]["Name"].tolist()
h_q_c = df[(df["Completeness"] >= 90) & df["Contamination"].between(5, 10)]["Name"].tolist()
m_q = df[(df["Completeness"] >= 50) & (df["Completeness"] < 90) & (df["Contamination"] < 10)]["Name"].tolist()
l_q = df[(df["Completeness"] < 50) & (df["Contamination"] < 10)]["Name"].tolist()
cont = df[(df["Contamination"] > 10)]["Name"].tolist()
```

```python
print(len(h_q))
print(len(h_q_c))
print(len(m_q))
print(len(l_q))
print(len(cont))
```

```python
approved_bins = h_q + h_q_c + m_q

data = pd.read_csv("../chen_data_5.csv")
data.head()
data_qc = data[data['sample_name'].isin(approved_bins)]

data_qc.to_csv('../chen_data_qc.csv', index=False)
```
